# S&P500it index

### Testing with ADF, SADF, GSADF and BSADF

```python
# imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
```

##### 1. Download S&P 500 IT index from 1. may 1993 to 31. dec. 2025

```python
# Download S&P 500 Information Technology Sector, ticker = "^SP500-45" from yahoo finance
sp500it = yf.Ticker("^SP500-45")
sp500it_data = sp500it.history(start="1990-01-01", end="2025-12-31", interval="1mo")
sp500it_close = sp500it_data["Close"]
log_sp500it_close = np.log(sp500it_close)
```

```python
# Plot the S&P 500 Information Technology Sector index
plt.plot(sp500it_close.index, sp500it_close, color="C1", ls="-", label="S&P 500 Information Technology")
plt.grid(True, alpha=0.7, ls="--")
plt.legend()
plt.savefig("plots/sp500it.pdf", bbox_inches="tight")
plt.show()
```

```python
# Plot the logarithmic S&P 500 Information Technology Sector index
plt.plot(log_sp500it_close.index, log_sp500it_close, color="C1", ls="-", label="S&P 500 Information Technology")
plt.grid(True, alpha=0.7, ls="--") 
plt.legend()
plt.savefig("plots/sp500it_log.pdf", bbox_inches="tight")
plt.show()
```

```python
# download log and close as .csv
sp500it_close.to_csv("data/sp500it.csv")
log_sp500it_close.to_csv("data/log_sp500it.csv")
```

##### 2. We now display results of testing S&P 500 IT index from 1. may 1993 to 31. dec. 2025
Results on untransformed prices, $P_t$:

BSADF with untransformed prices, $P_t$:

```python
# download bsadf_stat.csv and bsadf_cv.csv
bsadf_stat = pd.read_csv("results/sp500it_bsadf_stat.csv")
bsadf_cv   = pd.read_csv("results/sp500it_bsadf_cv.csv")

t = sp500it_close.index[-len(bsadf_stat):]

plt.figure(figsize=(8, 4))
plt.plot(t, bsadf_stat["bsadf"], label="BSADF statistic")
plt.plot(t, bsadf_cv["95%"], "--", label="95% critical value")
plt.plot(t, bsadf_cv["99%"], "--", label="99% critical value")
plt.legend()
plt.grid(True, alpha=0.7, ls="--")
plt.show()
```

Results on transformed prices, $\ln(P_t)$:

BSADF with transformed prices, $\ln(P_t)$:

```python
bsadf_stat = pd.read_csv("results/log_sp500it_bsadf_stat.csv")
bsadf_cv   = pd.read_csv("results/log_sp500it_bsadf_cv.csv")

t = log_sp500it_close.index[-len(bsadf_stat):]

plt.figure(figsize=(8, 4))
plt.plot(t, bsadf_stat["bsadf"], label="BSADF statistic")
plt.plot(t, bsadf_cv["95%"], "--", label="95% critical value")
plt.plot(t, bsadf_cv["99%"], "--", label="99% critical value")
plt.legend()
plt.grid(True, alpha=0.7, ls="--")
plt.show()
```
