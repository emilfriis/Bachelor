```python
import matplotlib.pyplot as plt
import itertools

# Windows for ADF tests
windows = [
    (0.0, 0.1),
    (0.0, 0.2),
    (0.0, 0.3),
    (0.0, 0.4),
    (0.0, 0.5),
    (0.1, 0.2),
    (0.1, 0.3),
    (0.1, 0.4),
    (0.1, 0.5),
    (0.2, 0.3),
    (0.2, 0.4),
    (0.2, 0.5),
    (0.3, 0.4),
    (0.3, 0.5),
    (0.4, 0.5),
]

fig, ax = plt.subplots(figsize=(6, 4))

# map fra startpunkt -> farve
color_cycle = itertools.cycle(plt.cm.tab10.colors)
color_for_start = {}

for i, (start, end) in enumerate(windows):
    if start not in color_for_start:
        color_for_start[start] = next(color_cycle)
    c = color_for_start[start]

    y = -i  # forskudt nedad pr. vindue
    if i == 0:
        ax.hlines(y, start, end, linewidth=1.5, color=c,
                  label="Test windows in the GSADF")
    else:
        ax.hlines(y, start, end, linewidth=1.5, color=c)

    # pilspids ved højre ende
    ax.plot(end, y, marker=">", color=c, markersize=4, zorder=10)

# lodrette stiplede linjer
# for x in [1, 2, 3, 4, 5]:
#     ax.axvline(x, linestyle="--", alpha=1, color="black", lw=0.8)

ax.set_ylim(-len(windows)-0.5, 0.5)
ax.set_xlabel("$r_0$")
ax.set_yticks([])  # skjul y-akse
plt.legend(loc="lower left")
plt.tight_layout()
plt.grid(True)
plt.show()
```
